```python
import json
```

```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
```

```python
import re
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords 
```

```python
captions=[]
```

```python
with open("./captions_train2014.json", "r") as st_json:
    data = json.load(st_json)
    
for d in data['annotations']:
    captions.append(d['caption'].rstrip())
```

```python
with open("./captions_val2014.json", "r") as st_json:
    data = json.load(st_json)
    
for d in data['annotations']:
    captions.append(d['caption'].rstrip())
```

```python
with open("./captions_val2017.json", "r") as st_json:
    data = json.load(st_json)
for d in data['annotations']:
    captions.append(d['caption'].rstrip())
```

```python
len(captions)
```

```python
captions
```

```python
preprocess = []
for s in captions:
    tokens = re.sub(r"[^a-z0-9]+", " ", s.lower())
    preprocess.append(tokens)
```

```python
preprocess
```

```python
token = [word_tokenize(s) for s in preprocess]
```

```python
stop_words = set(stopwords.words('english')) 
```

```python
token
```

```python
text = []
for s in token:
    tmp=[]
    for w in s: 
        if w not in stop_words: 
            tmp.append(w) 
    text.append(tmp)
```

```python
text
```

```python
model = Word2Vec(sentences=token, size=64, window=5, min_count=5, workers=4, sg=0)
```
